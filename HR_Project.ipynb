```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('workingdata_HR.csv')
print(data)
```

```python
# Selecting relevant columns for correlation analysis
sustainable_performance = data.iloc[:, 7:22]  # Columns H to V
hrm = data.iloc[:, 22:44]                      # Columns W to AR
employee_behavior = data.iloc[:, 44:49]       # Columns AS to AW
culture = data.iloc[:, 49:55]                  # Columns AX to BC
self_efficacy = data.iloc[:, 55:61]           # Columns BD to BI
environmental_performance = data.iloc[:, 61:65] # Columns BJ to BM
```

```python
# Concatenating all relevant variables into one DataFrame for correlation
analysis_data = pd.concat([sustainable_performance, hrm, employee_behavior, culture, self_efficacy, environmental_performance], axis=1)
```

```python
# Calculating the correlation matrix
analysis_data_1 = pd.concat([employee_behavior, environmental_performance], axis=1)
correlation_matrix = analysis_data_1.corr()
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Heatmap of Employee Behavior and Environmental Performance')
plt.show()
```

```python
# Calculating the correlation matrix
analysis_data = pd.concat([sustainable_performance, culture], axis=1)
correlation_matrix = analysis_data.corr()
# Set up the matplotlib figure
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Heatmap of Sustainable performance and culture')
plt.show()
```

```python
# Calculating the correlation matrix
analysis_data = pd.concat([sustainable_performance, hrm, employee_behavior, culture, self_efficacy, environmental_performance], axis=1)
correlation_matrix_all = analysis_data.corr()
# Flatten the correlation matrix and exclude self-correlations (diagonal)
correlation_values = correlation_matrix_all.stack()

# Get the maximum correlation value and the variables involved
max_correlation = correlation_values[correlation_values != 1].max()
max_correlation_vars = correlation_values[correlation_values == max_correlation].index

# Get the minimum correlation value and the variables involved
min_correlation = correlation_values.min()
min_correlation_vars = correlation_values[correlation_values == min_correlation].index

# Print the results
print("Maximum correlation value:", max_correlation)
print("Variables with maximum correlation:", max_correlation_vars)
print("Minimum correlation value:", min_correlation)
print("Variables with minimum correlation:", min_correlation_vars)
```
